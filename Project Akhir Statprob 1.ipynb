```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd as xl

df = pd.read_excel("STATPROB 1.xlsx")
df
```

```python
slices = df['Berapa Lama Kalian Mengerjakan Tugas'].value_counts()
explode = [0.1, 0.1, 0.1]
label = ["2-3 Jam", "4-5 Jam", ">5 Jam"]
plt.gca().set_facecolor('#f0f0f0') 
plt.pie(slices, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90, )
plt.title("Jam Belajar Mahasiswa Ilkom")
plt.legend(label, loc=3)
plt.tight_layout()
plt.show()
```

```python
slices = df['Apakah dengan Adanya Tugas Mempengaruhi Pola Tidur dan Pola Makan?'].value_counts()
explode = [0.1, 0.1]
labels = [f"{label} ({count})" for label, count in zip(slices.index, slices.values)]
plt.pie(slices, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90, )
plt.title("Pengaruh ke Pola Makan dan Pola Tidur Mahasiswa Ilkom")
plt.legend(labels, loc=3)
plt.tight_layout()
plt.show()
```

```python
cost_ranges = [
    'Tidak Berpengaruh',
    'Cukup Berpengaruh',
    'Berpengaruh',
    'Sangat Berpengaruh'
]

data = df['Apakah dengan Adanya Tugas Mempengaruhi Kesehatan Fisik Anda?'].value_counts()
colors = ['#a6cee3', '#1f78b4', '#b2df8a', '#33a02c']
plt.figure(figsize=(10, 6))
plt.bar(cost_ranges, data, color=colors, edgecolor='black' ,width=0.5)
plt.xticks(rotation=45, ha='right')  
plt.xlabel('Rentang Pengeluaran Bensin', fontweight='bold', fontsize=15)
plt.ylabel('Frekuensi', fontweight='bold', fontsize=15) 
plt.grid(True, alpha=0.7, color='gray')
plt.title('Distribusi Pengeluaran', fontweight='bold', fontsize=20)
```

```python
slices = df['Apakah Tugas yang Diberikan oleh Dosen Bermanfaat dalam Pemahaman Terhadap Mata Kuliah?'].value_counts()
explode = [0.1, 0.1]
label = ["Ya", "Tidak"]
plt.gca().set_facecolor('#f0f0f0') 
plt.pie(slices, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90, )
plt.title("Apakah Bermanfaat dalam Pemahaman Mahasiswa Ilkom")
plt.legend(label, loc=3)
plt.tight_layout()
plt.show()
```
