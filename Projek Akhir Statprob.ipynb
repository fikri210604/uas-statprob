```python
data = {
    'Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?': [np.nan, np.nan, np.nan, '40.500-50.000', '30.500-40.000', np.nan, '80.500-90.000', '50.500-60.000', '20.000-30.000', '30.500-40.000'],
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Pra-pemrosesan data: Konversi rentang menjadi nilai tengah
def process_range(value):
    if pd.isna(value):
        return np.nan
    elif '-' in value:
        range_values = list(map(float, value.replace('-', '').split('-')))
        return np.mean(range_values)
    else:
        return float(value)

df['Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?'] = df['Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?'].apply(process_range)

# Menampilkan DataFrame setelah pra-pemrosesan
print(df)

# Pendugaan parameter
mean_value = df['Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?'].mean()
median_value = df['Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?'].median()
std_dev = df['Berapakah Pengeluaran untuk Bensin Kendaraan Anda dalam Seminggu?'].std()

print(f"Rata-rata: {mean_value}")
print(f"Median: {median_value}")
print(f"Simpangan Baku: {std_dev}")
```
