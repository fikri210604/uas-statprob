```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd as xl
import seaborn as sns

df = pd.read_csv("DATA Transportasi - Form Responses 1.csv")
df
```

```python
slices = df['Dengan Apa Anda Pergi ke Unila'].value_counts()
explode = [0.1, 0.1]
labels = [f"{label} ({count})" for label, count in zip(slices.index, slices.values)]
plt.pie(slices, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Dengan Apa Anda Pergi ke Unila", fontweight="bold", fontsize=20)
plt.legend(labels, loc=3, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
df['Bensin'] = pd.to_numeric(df['Bensin'], errors='coerce')
nilai_terkecil = df['Bensin'].min()
nilai_terbesar = df['Bensin'].max()
jangkauan = df['Bensin'].max() - df['Bensin'].min()

# Menampilkan hasil
print("Nilai Terkecil Bensin:", nilai_terkecil)
print("Nilai Terbesar Bensin:", nilai_terbesar)
print("Jangkauan Bensin:", jangkauan)
# Menampilkan statistik
print("Mean Bensin:", df['Bensin'].mean())
print("Median Bensin:", df['Bensin'].median())
print("Std Dev Bensin:", df['Bensin'].std())
```

```python

# Mengganti nilai kosong dengan NaN
df['Bensin'] = pd.to_numeric(df['Bensin'], errors='coerce')

# Membuat subplot untuk histogram Bensin
fig, ax = plt.subplots(figsize=(8, 5))

# Histogram Bensin
sns.histplot(df['Bensin'].dropna(), kde=True, color='#1f78b4', edgecolor='black', ax=ax)
ax.set_title('Histogram Bensin')
ax.set_xlabel('Bensin')
ax.set_ylabel('Frekuensi')

# Menampilkan plot
plt.show()

```

```python
df['Bensin'] = pd.to_numeric(df['Bensin'], errors='coerce')

# Membuat subplot untuk boxplot Bensin


# Boxplot Bensin
sns.boxplot(y=df['Bensin'].dropna(), color='lightgreen', orient='v', ax=ax)
ax.set_title('Boxplot Bensin')
ax.set_ylabel('Bensin')

# Menampilkan plot
plt.show()
```

```python
slices = df['Alat'].value_counts()
labels = [f"{label} ({count})" for label, count in zip(slices.index, slices.values)]
explode = [0.1] * len(labels) 
plt.pie(slices, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Dengan Apa Anda Pergi ke Unila", loc='center', pad=20, fontsize=16, fontweight='bold', color='blue', fontstyle='normal')
plt.legend(labels, loc=3, fontsize = 6)
plt.show()
```
