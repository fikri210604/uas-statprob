```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xlrd as xl
import openpyxl as op

df = pd.read_excel(r'C:\Users\Administrator\Documents\tugas-kuliah\EKSPOR KELAPA SAWIT INDONESIA BERDASARKAN NEGARA TAHUN 2012-2022.xls')
df
```

```python

x = df['Negara tujuan']
y = df['2022']  

# Buat warna yang berbeda untuk setiap bar
colors = ['red', 'blue', 'green', 'orange','purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# Plot bar chart
plt.figure(figsize=(20,10))
plt.bar(x, y, color=colors)

# Label dan judul grafik
plt.xlabel('Negara Tujuan', fontsize=15)
plt.ylabel('Volume Ekspor (per 1000 Ton)', fontsize=15)
plt.title('Tujuan Ekspor Berdasarkan Negara Tahun 2022', fontsize=20)

# Rotasi label sumbu x
plt.xticks(rotation=45, ha='right', fontsize=10)

# Batasan sumbu y
plt.yticks(range(0, 5500, 500))

# Warna grid
plt.grid(True, alpha=0.7, color='gray')

# Menampilkan grafik
plt.show()
```

```python
afh = pd.read_excel(r'C:\Users\Administrator\Documents\tugas-kuliah\Book1.xlsx')

tahun = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
india = afh['India']
tiongkok = afh['Tiongkok']
pakistan = afh['Pakistan']
belanda = afh['Belanda']
amerika = afh['Amerika Serikat']
spanyol = afh['Spanyol']
mesir = afh['Mesir']
bangladesh = afh['Bangladesh']
italia = afh['Italia']
singapura = afh['Singapura']
```

```python
fig = plt.figure()
myaxes = fig.add_axes([0,0,1,1])
plt.title('Jumlah Ekspor Kelapa Sawit Menurut Negara 2012-2022', fontsize=10)
plt.xlabel('Tahun')
plt.ylabel('Volume Ekspor (per 1000 Ton)')
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(range(0, 10000, 1000))
plt.gridcolor = "Lightgreen"
plt.figure(figsize=(20,10))
myaxes.plot(tahun, singapura, marker='o')
myaxes.plot(tahun, india, marker='o')
myaxes.plot(tahun, tiongkok, marker='o')
myaxes.plot(tahun, pakistan, marker='o')
myaxes.plot(tahun, belanda, marker='o')
myaxes.plot(tahun, amerika, marker='o')
myaxes.plot(tahun, spanyol, marker='o')
myaxes.plot(tahun, mesir, marker='o')
myaxes.plot(tahun, bangladesh, marker='o')
myaxes.plot(tahun, italia, marker='o')

myaxes.legend(['Singapura', 'India', 'Tiongkok', 'Pakistan', 'Belanda', 'Amerika Serikat', 'Spanyol', 'Mesir', 'Bangladesh', 'Italia'], loc='upper right',fontsize=6)
```

```python

Total = afh['Total Ekspor']
jumlah =afh['Nilai Ekspor']
fig = plt.figure()
myaxes = fig.add_axes([0,0,1,1])
plt.grid(True)
plt.xticks(rotation=45)
plt.title('Perbandingan Total dan Nilai Ekspor Kelapa Sawit 2012-2022', fontsize=10)
plt.xlabel('Tahun')
plt.ylabel('Volume Ekspor (per 1000 Ton)')
plt.gridcolor = "Lightgreen"
plt.figure(figsize=(20,10))
myaxes.plot(tahun, Total, marker='o', color='red',linestyle=':',linewidth=2)
myaxes.plot(tahun, jumlah, marker='o', color='blue',linestyle='--',linewidth=2)
myaxes.legend(['Total Ekspor', 'Nilai Ekspor'], loc='upper right',fontsize=10)
plt.show()
```
