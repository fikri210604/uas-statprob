```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd as xl
import seaborn as sns

df = pd.read_csv("berapa-konsumsi-bensin-ron-88-dan-ron-90-di-indonesia.csv")
df
```

```python

# Mengatur kolom 'Tanggal' sebagai indeks
df.set_index('date', inplace=True)

# Membuat plot data time series
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Bensin RON 88'], marker='o', linestyle='-')
plt.plot(df.index, df['Bensin RON 90'], marker='o', linestyle='-')
plt.plot(df.index, df['Bensin RON 92'], marker='o', linestyle='-')

# Menambahkan judul dan label sumbu
plt.title('Plot Data Time Series')
plt.xlabel('Tanggal')
plt.ylabel('Nilai')

# Menampilkan grid
plt.grid(True)
plt.legend(['Bensin RON 88', 'Bensin RON 90', 'Bensin RON 92'])

# Menampilkan plot
plt.show()
```
